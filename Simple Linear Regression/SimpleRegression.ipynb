```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2 , random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regessor = LinearRegression()
regessor.fit(X_train,y_train)
```

```python
y_pred = regessor.predict(X_test)
```

```python
#Visualize traininng results
plt.scatter(X_train, y_train , color= 'red')
plt.plot(X_train ,regessor.predict(X_train) , color = 'blue')
plt.title('Regression result with input data')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualze testing results
plt.scatter(X_test, y_test , color= 'red')
plt.plot(X_train ,regessor.predict(X_train) , color = 'blue')
plt.title('Regression result with output data')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(np.rint(y_test),np.rint(y_pred))
#np.rint(y_test)
```

```python
import seaborn as sns
sns.heatmap(cm)
```
